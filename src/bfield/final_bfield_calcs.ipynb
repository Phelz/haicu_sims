```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

import data
```

```python
bfield_bender = pd.read_csv(data.DIR_PATH / 'bfield' / 'FA_Bfield_Bender_Iso_100large.csv')
bfield_qpole = pd.read_csv(data.DIR_PATH / 'bfield' / 'FA_Bfield_Qpole_Iso_100large.csv')
bfield_outergatecoil = pd.read_csv(data.DIR_PATH / 'bfield' / 'FA_Bfield_OuterGateBCoil_Iso_100large.csv')
```

```python
total_field = bfield_bender.copy()
total_field['Bx'] = bfield_bender['Bx'] + bfield_qpole['Bx'] + bfield_outergatecoil['Bx']
total_field['By'] = bfield_bender['By'] + bfield_qpole['By'] + bfield_outergatecoil['By']
total_field['Bz'] = bfield_bender['Bz'] + bfield_qpole['Bz'] + bfield_outergatecoil['Bz']

total_field.to_csv(data.DIR_PATH / 'bfield' / 'FA_Bfield_total_Iso_100large.csv', index=False)
```

```python
# Number of unique values
print(total_field['x'].diff().unique())
print(total_field['y'].diff().unique())
print(total_field['z'].diff().unique())
```

```python
# Number of unique values
print(len(total_field['x'].unique()))
print(len(total_field['y'].unique()))
print(len(total_field['z'].unique()))
```

```python
# Now to calculate the field gradient
field_gradient = total_field.copy()
field_gradient['dBx_dz'] = np.gradient(total_field['Bx'], total_field['z'])
field_gradient['dBy_dz'] = np.gradient(total_field['By'], total_field['z'])
field_gradient['dBz_dz'] = np.gradient(total_field['Bz'], total_field['z'])

field_gradient
```

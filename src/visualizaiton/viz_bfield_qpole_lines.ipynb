```python
from definitions import *

import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import data, figures
from mpl_stylesheets import STYLE_SHEETS_PATHS
plt.style.use(STYLE_SHEETS_PATHS)
```

```python
line_30deg_data = pd.read_csv(data.DIR_PATH / 'B_field_qpole_30deg_line.csv')
line_30deg_data['magitude'] = np.sqrt(line_30deg_data['Bx']**2 + line_30deg_data['By']**2 + line_30deg_data['Bz']**2)
line_30deg_data['r'] = np.sqrt(line_30deg_data['x']**2 + line_30deg_data['y']**2 + line_30deg_data['z']**2)

angles = [30, 45, 60, 75, 90]
# angles = [30, 45, 60, 75]
dfs = [
    pd.read_csv(data.DIR_PATH / f'B_field_qpole_{angle}deg_line.csv') for angle in angles
]

for df in dfs:
    df['magnitude'] = np.sqrt(df['Bx']**2 + df['By']**2 + df['Bz']**2)
    df['r'] = np.sqrt(df['x']**2 + df['y']**2 + df['z']**2)
```

```python

# make a plot of r vs magnitude


# Use a rainbow of colors for the lines
colors = [
    '#000000',  # Red
    '#FF0000',  # Red
    '#FF7F00',  # Orange
    '#00FFFF',  # Cyan
    '#0000FF',  # Blue
    '#7F00FF',  # Purple
]

fig, ax = plt.subplots(figsize=(15, 10))

for df, angle, color in zip(dfs[::-1], angles[::-1], colors):
    ax.plot(df['r']*1000, df['magnitude']*1000, label=f'{angle}Â°', ms=0, lw=1, alpha=1, color=color, zorder=10)
    
    
ax.set_xlabel('r (mm)')
ax.set_ylabel('B-field magnitude (mT)')

ax.grid(ls='--', lw=0.5, alpha=0.5)

ax.set_xlim(0, 50)
ax.set_ylim(0, 770)
ax.set_xticks(np.arange(0, 51, 10))
ax.set_yticks(np.arange(0, 751, 50))
ax.legend(fontsize=18)
```
